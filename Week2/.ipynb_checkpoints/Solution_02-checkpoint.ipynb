### Pacchetti

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
import scipy.optimize
import math
```

### Funzioni

```python
## pdf gaussiana
def gauss (x, mean=0, std=1):
    return (1./(np.sqrt(2*math.pi)*std)) * np.exp((-1./2.)*((x-mean)**2)/std**2)

## pdf esponenziale
def expon (x, lamda=1, base=math.e):
    return lamda*np.log(base)*base**(-lamda*x)

## pdf Cauchy
def cauchy (x, gamma=1, mu=0):
    return (1./math.pi)*gamma/( (x-mu)**2 + gamma**2 )

## sigma
def sigma (x, a=1):
    return a*np.sqrt(x)

## altezze degli istogrammi per fit
def get_heights (df, bins=101, density=True):
    
    fig, ax = plt.subplots()
    n, bins_new, rects = ax.hist(df, bins=bins, density=density)
    plt.close()

    x = bins_new[:-1]
    y = []
    for r in rects:
        y.append(r.get_height())
        
    return np.array(y)

## fit della pdf
def fit_graph (df, graph_col, title, ylabel, xlabel=r"S$_{100}$", x_range=None, p0=None, density=True, bins=101,
               fontsize=15, g_bool=False, c_bool=False):
    
    df = df[graph_col].dropna()
    if x_range != None:
        x = np.linspace(x_range[0], x_range[1], bins)
        df = df[df > x_range[0]][df < x_range[1]]
    else:
        x = np.linspace(df.min(), df.max(), bins)
    
    
    plt.rcParams.update({'font.size': fontsize})
    fig, ax = plt.subplots(figsize=(8, 4))
    fig.suptitle(title)

    ax.hist(df, bins=bins, density=density, label='data')
    
    if g_bool:
        y = get_heights(df=df, bins=bins)
        pars_val, pars_cov = scipy.optimize.curve_fit(gauss, x, y, p0=p0)
        #Gauss = scipy.stats.norm(loc=pars_op[0], scale=pars_op[1])
        #ax.plot(x, Gauss.pdf(x), color='red', linestyle='-', label='fit (Gauss)')
        ax.plot(x, gauss(x, pars_val[0], pars_val[1]), color='red', linestyle='-', label='fit (Gauss)')
    if c_bool:
        y = get_heights(df=df, bins=bins)
        pars_val, pars_cov = scipy.optimize.curve_fit(cauchy, x, y, p0=p0)
        ax.plot(x, cauchy(x, pars_val[0], pars_val[1]), color='red', linestyle='-', label='fit (Cauchy)')
        
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend()
    ax.grid()
```

### Esercizio 02.1

```python
## caricamento dati
df_media = pd.read_csv("./data_02.1/media_data.txt", sep="\t")
df_sample = pd.read_csv("./data_02.1/sample_data.txt", sep="\t")
df_ar = pd.read_csv("./data_02.1/ar_data.txt", sep="\t")
```

```python
## METODO DELLA MEDIA (plot della media progressiva con errore statistico)
x = np.arange(len(df_media)) * 1000
mean_teo = np.zeros(len(df_media)) + 1.

plt.rcParams.update({'font.size': 13})
fig, ax = plt.subplots(figsize=(8, 4))
fig.suptitle("Media progressiva ed errore statistico")

ax.errorbar(x, df_media.iloc[:, 0], yerr=df_media.iloc[:, 1], color="b", label="Media")
ax.errorbar(x, df_sample.iloc[:, 0], yerr=df_sample.iloc[:, 1], color="y", label="Sample")
ax.errorbar(x, df_ar.iloc[:, 0], yerr=df_ar.iloc[:, 1], color="g", label="AR")
ax.plot(x, mean_teo, color='r', linestyle='-')

ax.set_xlabel("#integrali")
ax.set_ylabel("<r>")
#ax.set_ylim([0.998, 1.002])
ax.grid(True)
ax.legend();
```

### Esercizio 02.2

```python
## caricamento dati
df_discrete = pd.read_csv("./data_02.2/discrete.txt", sep="\t")
df_continous = pd.read_csv("./data_02.2/continous.txt", sep="\t")
dfs = [df_discrete, df_continous]
```

```python
## plot dell'andamento di sigma per un rw discreto
x = np.arange(len(df_discrete))
a = 3 #passo del rw
colors = ["green", "blue"]
labels = ["Discreto", "Continuo"]

plt.rcParams.update({'font.size': 13})
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, sharey=False, figsize=(16, 4))
fig.suptitle(r"Andamento di $\sigma$$_N$")

axes = [ax1, ax2]
for ax in axes:
    i = axes.index(ax)
    ax.errorbar(x, dfs[i].iloc[:, 0], yerr=dfs[i].iloc[:, 1], color=colors[i], label=labels[i])
    ax.plot(x, sigma(x, a), color="red", label="Teorico")

    ax.set_xlabel("N")
    ax.set_ylabel(r"$\sigma$$_N$")
    ax.grid(True)
    ax.legend();
```

### Abstract
In this lecture, the problem is to perform a classification task, using Deep Neural Network models implemented in the Keras python package, to recognize and distinguish between the ten handwritten digits (0-9).

We will get our training and test sets from the MNIST dataset, which comprises $70000$ handwritten digits, each of which comes in a square image, divided into a $28\times 28$ pixel grid. Every pixel can take on $256$ gradation of the gray color, interpolating between white and black, and hence each data point assumes any value in the set $\{0,1,\dots,255\}$. For this reason, the pixels values will be renormalized to $[0, 1]$ before the training session.

The goal is to study the different features of DNNs (Deep Neaural Networks), CNNs (Convolutional Neural Networks) and finally DA (Data Augmentation) models. In order to do so:
1. in the first exercise, we will build three DNN models with the same layer structue but using different optimizers, in order to find out which one is the best suited for this classification task;
2. then, in the second exercise, we will add on top of the previously defined DNN structure a couple of convolutional layers. These will help the model in recognizing the different features of each digit;
3. finally, in the last exercise, we will observe the lack of versatility of the previously trained DNN and CNN models and the greater versatility of DA models.

---
Each model has been provided with:
- loss function: Sparse Categorical Crossentropy
- optimizer: Adam;
- batch size: 32;
- validation split: 0.3.

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
def Print_score(exercise):
    scores = {}
    with open("../data/ex1/test_metrics.dat", 'r') as fp:
        for iopt in range(3):
            optimizer = fp.readline().split("\n")[0]
            loss = float(fp.readline().split("\n")[0])
            accuracy = float(fp.readline().split("\n")[0])
            scores[optimizer] = [loss, accuracy]
    
    with open("../data/ex2/test_metrics.dat", 'r') as fp:
        loss = float(fp.readline().split("\n")[0])
        accuracy = float(fp.readline().split("\n")[0])
        CNN_score = [loss, accuracy]
        
    with open("../data/ex3/test_metrics.dat", 'r') as fp:
        loss = float(fp.readline().split("\n")[0])
        accuracy = float(fp.readline().split("\n")[0])
        DA_score = [loss, accuracy]
    
    if exercise == '01':
        print("DNN models metrics when evaluated on the test set:\n")
        for optimizer in scores:
            print(f"loss={format(scores[optimizer][0], '.5f')}, accuracy={format(scores[optimizer][1], '.5f')}  ({optimizer})")
        
    elif exercise == '02':
        print("CNN model metrics when evaluated on the test set:")
        print(f"loss={format(CNN_score[0], '.5f')}, accuracy={format(CNN_score[1], '.5f')}")
        print()
        print("Accuracy for:")
        print(f"Best DNN model (adam): {format(scores['adam'][1], '.5f')}")
        print(f"CNN model:             {format(CNN_score[1], '.5f')}")
        print(f"Difference:            {format(CNN_score[1] - scores['adam'][1], '.5f')}")
        
    elif exercise == '03':
        print("DA model metrics when evaluated on the test set:")
        print(f"loss={format(DA_score[0], '.5f')}, accuracy={format(DA_score[1], '.5f')}")
        print()
```

As anticipated, in this first exercise, we built a DNN structure, composed by:
- Dense layer, with 64 neurons and "relu" activation function;
- Dense layer, with 32 neurons and "relu" activation function;
- Dense layer, with 10 neurons and "softmax" activation function (as it is a classification problem).

As in the previous lecture, the training session has been stopped after 20 epochs without improvement on the validation set loss function (with a maximum of 100 epochs).

We should make a couple of observations:
- first of all, the importance of the validation set in preventing the model from overfitting (as it can be seen in the loss function shape for the validation set throughout the epochs);
- moreover, despite the different optimizers, each model reaches a high accuracy percentage, $\approx 97\%$, when evaluated on the validation set during the training.

We want to now know how these models perform on the test set and which of them is the best one.

```python
Print_score('01')
```

We can see that the best model (accuracy-wise) is the one provided with the Adam optimizer. It should also be noted that the other models don't differ much (again, accuracy-wise) from the best one.

```python
Image(filename='../data/ex2/metrics.png')
```

In this second exercise, we added two convolutional layers to the previous DNN structure:
- Conv2D layer, with 32 filters, (3, 3) as kernel size and 'relu' activation function + MaxPool2D layer with (2, 2) as pool size;
- Conv2D layer, with 16 filters, (3, 3) as kernel size and 'relu' activation function + MaxPool2D layer with (2, 2) as pool size.

Again, the training session has been stopped after 20 epochs without improvement on the validation set loss function (with a maximum of 100 epochs).

```python
Print_score('02')
```

We can observe that the addition of the convolutional layer brought a improvement in accuracy of $\approx 1.7\%$, which is pretty significative considering the fact that the DNN models precision was already extremely high. This shows the utility of feature extraction provided by convolutional layers.

### Exercise 12.03

For this exercise, I generated two sets of 10 "handwritten" digits using the "gimp" application, each of which represented a number from 0 to 9. I then evaluated the (best) DNN model and the CNN on these two test sets.

```python
Image(filename='../data/ex3/first_normal_prediction.png')
```

The first test set is composed by pictures which are similar to the set used to perform the training of the two models. As expected, the CNN model performs better than the DNN one, which still correctly classificates 8 picture out of 10.

```python
Image(filename='../data/ex3/first_weird_prediction.png')
```

This second set is instead composed by rotated and distorted pictures of the previous ten digits. In this case, we can see that both models clearly fail to distinguish most of the digits. Again, this is the same problem encountered in Lecture 11, where we saw the regression problem for a DNN algorithm on a polynomial function.

```python
Image(filename='../data/ex3/metrics.png')
```

To try and fix this problem (that is the poor adaptability of the DNN and CNN models), we can build a DA model, adding on top of the previous CNN model some layers which will expand our training set by distorting, zooming, rotating and translating the pictures obtained from the MNIST dataset. These added layers are:
- RandomRotation layer, with 0.3 rotation factor;
- RandomZoom layer, with 0.2 height factor;
- RandomTranslation, with height factor $\in (-0.4, 0.4)$ and width factor $\in (-0.4, 0.4)$.

Once again, the training session has been stopped after 20 epochs without improvement on the validation set loss function (with a maximum of 100 epochs).

```python
Print_score('03')
```

As expected, the accuracy is not as high as the previous model. This is due to the fact that the dataset has been highly diversified, thus it is much more difficult for the model to extract generic digit features from the images. However, it should be noted that the accuracy is still around $\approx 80\%$.

```python
Image(filename='../data/ex3/second_normal_prediction.png')
```

When tested on the "normal" images, the DA model scores worse than any of the other two, as already seen from the test set accuracy evaluation.

```python
Image(filename='../data/ex3/second_weird_prediction.png')
```

Instead, once the "weird" images are used, the DA model proves to be the most versatile, making 8 correct predictions out of 10 pictures. This shows that, at the cost of accuracy percentage, the CNN model equipped with Data Augmentation layers proves to be the most adaptive to new inputs.

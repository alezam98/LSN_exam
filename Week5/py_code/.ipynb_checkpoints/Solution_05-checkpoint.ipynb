### Abstract
In this lecture, we want to estimate the expectation value for the distance from the origin of the electron in a hydrogen atom, both in the ground state 1s and in the excited state 2p. To do so, we can use the Metropolis algorithm, with both an uniform transition probability and a gaussian one. The sampled positions (in Cartesian coordinates) can be used to estimate $\left\langle r \right\rangle$ and and we expect the following results: 
$$ 
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0
$$
where $a_0$ is the Bohr's radius.


Before starting the actual simulation, in order to use the Metropolis algorithm properly, we should discuss a few things.

First of all, we should determine the parameters values for the transition probabilities which would fix the acceptance rate to about $50\%$ . After some testing, the chosen values are $\delta_{1,0,0}=1.25$ and $\delta_{2,1,0}=0.75$, for uniform sampling and $\sigma_{1,0,0}=3.0$ and $\sigma_{2,1,0}=1.75$, for gaussian sampling.

Furthermore, since we want the blocks to be indipendent from one another, we should also fix the number of steps per block, $N_{steps}$. By looking at the autocorrelation function of the distance from the origin, $r$, it is clear that $N_{steps} \approx 100$ is sufficient to have small correlation between the starting and the ending position of the atom. The value used during the simulation is $N_{steps} = 1000$, well above the minimum quantity.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def radius(df):
    r = []
    for ipos in range(len(df)):
        r.append(np.sqrt(df["x"][ipos]**2 + df["y"][ipos]**2 + df["z"][ipos]**2))
    return r
```

### Exercise 05.1

```python
eq_config_dfs = {
    "gs_uniform": pd.read_csv("../data/ground_state/eq_configs_uniform.dat", sep="\t"),
    "1e_uniform": pd.read_csv("../data/excited_state/eq_configs_uniform.dat", sep="\t"),
    "gs_gauss": pd.read_csv("../data/ground_state/eq_configs_gaussian.dat", sep="\t"),
    "1e_gauss": pd.read_csv("../data/excited_state/eq_configs_gaussian.dat", sep="\t"),
}

config_dfs = {
    "gs_uniform": pd.read_csv("../data/ground_state/configs_uniform.dat", sep="\t"),
    "1e_uniform": pd.read_csv("../data/excited_state/configs_uniform.dat", sep="\t"),
    "gs_gauss": pd.read_csv("../data/ground_state/configs_gaussian.dat", sep="\t"),
    "1e_gauss": pd.read_csv("../data/excited_state/configs_gaussian.dat", sep="\t"),
}

mean_dfs = {
    "gs_uniform": pd.read_csv("../data/ground_state/r_uniform.dat", sep="\t"),
    "1e_uniform": pd.read_csv("../data/excited_state/r_uniform.dat", sep="\t"),
    "gs_gauss": pd.read_csv("../data/ground_state/r_gaussian.dat", sep="\t"),
    "1e_gauss": pd.read_csv("../data/excited_state/r_gaussian.dat", sep="\t"),
}
```

```python
labels = ["ground state 1s", "excited state 2p"]
states = ["gs_uniform", "1e_uniform"]
means = [1.5, 5.]

plt.rcParams.update({'font.size':15})
fig = plt.figure(figsize=(20, 18))

for i in range(2):
    
    state = states[i]
    label = labels[i]
    mean = means[i]
    
    df = eq_config_dfs[state]
    r = radius(df)
    steps = range(len(r))
    
    
    ax = plt.subplot(2, 2, i+1, projection="3d")
    ax.scatter(df["x"], df["y"], df["z"], c=df["z"], marker=".")
    ax.set_title("Equilibration of the "+label, size=16)
    ax.set_xlabel(r'x/a$_0$')
    ax.set_ylabel(r'y/a$_0$')
    ax.set_zlabel(r'z/a$_0$')
    ax.view_init(0, 0)

    ax = plt.subplot(2, 2, i+3)
    ax.plot(steps, r, color = "blue")
    ax.plot([0, len(r)], [mean, mean], color="red", label=r'<r>/a$_0$, '+label)
    ax.set_xlabel(r'steps')
    ax.set_ylabel(r'r/a$_0$')
    ax.legend()
    ax.grid(True)
```

Starting the simulation, the first step is making sure that the system is close to equilibrium. We can visualize what happens if we start from a point far away in the following graphs, where the starting point is chosen to be $(20, 20, 20)$ for the ground state equilibration and $(30, 30, 30)$ for the excited state one.

It is clear that, for both states, approximately $250$ steps are sufficient for the system to equilibrate. We can now start to sample the position of the system.

```python
blocks = np.arange(len(mean_dfs["gs_uniform"])) + 1
mean_gs = 1.5

plt.rcParams.update({'font.size':12})
fig, ax = plt.subplots(figsize=(10, 5))

ax.set_title("Ground state 1s")
ax.errorbar(blocks, mean_dfs["gs_uniform"]["Mean"], yerr=mean_dfs["gs_uniform"]["Error"], color="b", label="Uniform")
ax.errorbar(blocks, mean_dfs["gs_gauss"]["Mean"], yerr=mean_dfs["gs_gauss"]["Error"], color="g", label="Gauss")
ax.plot([blocks.min(), blocks.max()], [mean_gs, mean_gs], color="r", label="Teorical")
ax.set_xlabel("blocks")
ax.set_ylabel(r"<r>/a$_0$")
ax.legend()
ax.grid(True)

plt.show()


print("On the last block")
print("(uniform sampling)   estimate: ", format(mean_dfs["gs_uniform"].iloc[-1, 0], '.5f'), " ± ", format(mean_dfs["gs_uniform"].iloc[-1, 1], '.5f'), ",\texpected value: ", format(mean_gs, '.5f'))
print("(gaussian sampling)  estimate: ", format(mean_dfs["gs_gauss"].iloc[-1, 0], '.5f'), " ± ", format(mean_dfs["gs_gauss"].iloc[-1, 1], '.5f'), ",\texpected value: ", format(mean_gs, '.5f'))
```

We can observe that the estimates are compatible with the exact results within the respective statistical errors. There is no significative difference between uniform and gaussian sampling.

```python
blocks = np.arange(len(mean_dfs["1e_uniform"])) + 1
mean_1e = 5.

plt.rcParams.update({'font.size':12})
fig, ax = plt.subplots(figsize=(10, 5))

ax.set_title("Excited state 2p")
ax.errorbar(blocks, mean_dfs["1e_uniform"]["Mean"], yerr=mean_dfs["1e_uniform"]["Error"], color="b", label="Uniform")
ax.errorbar(blocks, mean_dfs["1e_gauss"]["Mean"], yerr=mean_dfs["1e_gauss"]["Error"], color="g", label="Gauss")
ax.plot([blocks.min(), blocks.max()], [mean_1e, mean_1e], color="r", label="Teorical")
ax.set_xlabel("blocks")
ax.set_ylabel(r"<r>/a$_0$")
ax.legend()
ax.grid(True)

plt.show()


print("On the last block")
print("(uniform sampling)   estimate: ", format(mean_dfs["1e_uniform"].iloc[-1, 0], '.5f'), " ± ", format(mean_dfs["1e_uniform"].iloc[-1, 1], '.5f'), ",\texpected value: ", format(mean_1e, '.5f'))
print("(gaussian sampling)  estimate: ", format(mean_dfs["1e_gauss"].iloc[-1, 0], '.5f'), " ± ", format(mean_dfs["1e_gauss"].iloc[-1, 1], '.5f'), ",\texpected value: ", format(mean_1e, '.5f'))
```

Again, we can observe that the estimates are compatible with the exact results within the respective statistical errors. There is no significative difference between uniform and gaussian sampling.

```python
plt.rcParams.update({'font.size':12})
fig = plt.figure(figsize=(20, 20))

for i in range(2):
    
    if i == 0:
        df_gs = config_dfs["gs_uniform"]
        df_1e = config_dfs["1e_uniform"]
        title = "Uniform"
        
    else:
        df_gs = config_dfs["gs_gauss"]
        df_1e = config_dfs["1e_gauss"]
        title = "Gaussian"
    
    # ground state
    ax = plt.subplot(2, 2, 2*i+1, projection="3d")
    ax.set_title(title+" sampling, Ground state 1s")
    ax.scatter(df_gs["x"], df_gs["y"], df_gs["z"], c=df_gs["z"], marker=".")
    ax.set_xlabel(r'x/a$_0$')
    ax.set_ylabel(r'y/a$_0$')
    ax.set_zlabel(r'z/a$_0$')
    ax.view_init(10, 30)

    # excited state
    ax = plt.subplot(2, 2, 2*i+2, projection="3d")
    ax.set_title(title+" sampling, Excited state 2p")
    ax.scatter(df_1e["x"], df_1e["y"], df_1e["z"], c=df_1e["z"], marker=".")
    ax.set_xlabel(r'x/a$_0$')
    ax.set_ylabel(r'y/a$_0$')
    ax.set_zlabel(r'z/a$_0$')
    ax.view_init(10, 30)
```

Lastly, pictures of the positions explored by the system during the execution of the Metropolis algorithm, divided by sampling method and quantum state. As before, there seems to be no significative difference between uniform and gaussian sampling.

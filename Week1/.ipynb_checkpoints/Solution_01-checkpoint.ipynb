### Pacchetti

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
import scipy.optimize
import math
```

### Funzioni

```python
## pdf gaussiana
def gauss (x, mean=0, std=1):
    return (1./(np.sqrt(2*math.pi)*std)) * np.exp((-1./2.)*((x-mean)**2)/std**2)

## pdf esponenziale
def expon (x, lamda=1, base=math.e):
    return lamda*np.log(base)*base**(-lamda*x)

## pdf Cauchy
def cauchy (x, gamma=1, mu=0):
    return (1./math.pi)*gamma/( (x-mu)**2 + gamma**2 )

## altezze degli istogrammi per fit
def get_heights (df, bins=101, density=True):
    
    fig, ax = plt.subplots()
    n, bins_new, rects = ax.hist(df, bins=bins, density=density)
    plt.close()

    x = bins_new[:-1]
    y = []
    for r in rects:
        y.append(r.get_height())
        
    return np.array(y)

## fit della pdf
def fit_graph (df, graph_col, title, ylabel, xlabel=r"S$_{100}$", x_range=None, p0=None, density=True, bins=101,
               fontsize=15, g_bool=False, c_bool=False):
    
    df = df[graph_col].dropna()
    if x_range != None:
        x = np.linspace(x_range[0], x_range[1], bins)
        df = df[df > x_range[0]][df < x_range[1]]
    else:
        x = np.linspace(df.min(), df.max(), bins)
    
    
    plt.rcParams.update({'font.size': fontsize})
    fig, ax = plt.subplots(figsize=(8, 4))
    fig.suptitle(title)

    ax.hist(df, bins=bins, density=density, label='data')
    
    if g_bool:
        y = get_heights(df=df, bins=bins)
        pars_val, pars_cov = scipy.optimize.curve_fit(gauss, x, y, p0=p0)
        #Gauss = scipy.stats.norm(loc=pars_op[0], scale=pars_op[1])
        #ax.plot(x, Gauss.pdf(x), color='red', linestyle='-', label='fit (Gauss)')
        ax.plot(x, gauss(x, pars_val[0], pars_val[1]), color='red', linestyle='-', label='fit (Gauss)')
    if c_bool:
        y = get_heights(df=df, bins=bins)
        pars_val, pars_cov = scipy.optimize.curve_fit(cauchy, x, y, p0=p0)
        ax.plot(x, cauchy(x, pars_val[0], pars_val[1]), color='red', linestyle='-', label='fit (Cauchy)')
        
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend()
    ax.grid()
```

### Esercizio 01.1

```python
## caricamento dati
df_mean = pd.read_csv("./data_01.1/mean_data.txt", sep="\t")
df_var = pd.read_csv("./data_01.1/var_data.txt", sep="\t")
df_chi = pd.read_csv("./data_01.1/chi_data.txt", sep="\t")
```

```python
## plot della media progressiva con errore statistico
x = np.arange(len(df_mean)) * 1000
mean_teo = np.zeros(len(df_mean)) + 1./2.

plt.rcParams.update({'font.size': 13})
fig, ax = plt.subplots(figsize=(8, 4))
fig.suptitle("Media progressiva ed errore statistico")
ax.errorbar(x, df_mean.iloc[:, 0], yerr=df_mean.iloc[:, 1])
ax.plot(x, mean_teo, color='r', linestyle='-')
ax.set_xlabel("#estrazioni")
ax.set_ylabel("<r>")
ax.grid(True)
```

```python
## plot della varianza progressiva con errore statistico
x = np.arange(len(df_mean)) * 1000
var_teo = np.zeros(len(df_mean)) + 1./12.

plt.rcParams.update({'font.size': 13})
fig, ax = plt.subplots(figsize=(8, 4))
fig.suptitle("Varianza media progressiva ed errore statistico")
ax.errorbar(x, df_var.iloc[:, 0], yerr=df_var.iloc[:, 1])
ax.plot(x, var_teo, color='r', linestyle='-')
ax.set_xlabel("#estrazioni")
ax.set_ylabel("<(r - 0.5)^2>")
ax.grid(True)
```

```python
## plot del chi quadro
x = np.arange(len(df_mean))

plt.rcParams.update({'font.size': 13})
fig, ax = plt.subplots(figsize=(8, 4))
fig.suptitle(r"$\chi^{2}$ in funzione del blocco")
ax.plot(x, df_chi.iloc[:, 0], linestyle='-')
ax.set_xlabel("blocco")
ax.set_ylabel(r'$\chi^{2}$')
ax.grid(True)
```

### Esercizio 01.2

```python
## caricamento dati
df_s1 = pd.read_csv("./data_01.2/Sampling_1.txt", sep="\t")
df_s2 = pd.read_csv("./data_01.2/Sampling_2.txt", sep="\t")
df_s10 = pd.read_csv("./data_01.2/Sampling_10.txt", sep="\t")
df_s100 = pd.read_csv("./data_01.2/Sampling_100.txt", sep="\t")
dfs = [df_s1, df_s2, df_s10, df_s100]

Ns = [1, 2, 10, 100]
```

```python
## distribuzione della media, pdf uniforme
pdf_orig = 'Uniform'
bins = np.linspace(1, 6, 101)

plt.rcParams.update({'font.size': 15})
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=False, figsize=(15, 8))
fig.suptitle(r"Distribuzione del valor medio S$_{N}$ per una distribuzione uniforme")
axs = [ax1, ax2, ax3, ax4]

for ax in axs:
    i = axs.index(ax)
    ax.hist(dfs[i][pdf_orig], bins, density=True)
    ax.set_xlabel(r"S$_{%d}$" % (Ns[i]))
    ax.grid(True)
    
fig.text(0.073, 0.5, 'pdf', va='center', rotation='vertical')
i=0
```

```python
## distribuzione della media, pdf esponenziale
pdf_orig = 'Exponential'
bins = np.linspace(0, 5, 101)

plt.rcParams.update({'font.size': 15})
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=False, figsize=(15, 8))
fig.suptitle(r"Distribuzione del valor medio S$_{N}$ per una distribuzione esponenziale")
axs = [ax1, ax2, ax3, ax4]

for ax in axs:
    i = axs.index(ax)
    ax.hist(dfs[i][pdf_orig], bins, density=True)
    ax.set_xlabel(r"S$_{%d}$" % (Ns[i]))
    ax.grid(True)
    
fig.text(0.073, 0.5, 'pdf', va='center', rotation='vertical')
i=0
```

```python
## distribuzione della media, pdf Cauchy
pdf_orig = 'Cauchy'
bins = np.linspace(-10, 10, 101)

plt.rcParams.update({'font.size': 15})
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True, figsize=(15, 8))
fig.suptitle(r"Distribuzione del valor medio S$_{N}$ per una distribuzione di Cauchy")
axs = [ax1, ax2, ax3, ax4]

for ax in axs:
    i = axs.index(ax)
    ax.hist(dfs[i][pdf_orig], bins, density=True)
    ax.set_xlabel(r"S$_{%d}$" % (Ns[i]))
    ax.grid(True)
    
fig.text(0.073, 0.5, 'pdf', va='center', rotation='vertical')
i=0
```

```python
## fit
graph_cols = df_s100.columns
titles = ["Fit da pdf uniforme", "Fit da pdf esponenziale", "Fit da pdf di Cauchy"]
ylabel = "pdf"
x_ranges = [[3, 4], [0.6, 1.4], [-10, 10]]
p0s = [[3.5, 0.5], [1., 0.4], [1., 0]]

for i in range(len(titles)):
    if i != 2:
        fit_graph (df=df_s100, graph_col=graph_cols[i], title=titles[i], ylabel=ylabel, x_range=x_ranges[i], 
                   p0=p0s[i], g_bool=True)
    else:
        fit_graph (df=df_s100, graph_col=graph_cols[i], title=titles[i], ylabel=ylabel, x_range=x_ranges[i], 
                   p0=p0s[i], c_bool=True)
```

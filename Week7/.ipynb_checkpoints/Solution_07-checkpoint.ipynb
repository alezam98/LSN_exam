### Pacchetti

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Funzioni

```python
def load_data(step_type):
    solid, liquid, gas = {}, {}, {}
    outputs = ["ekin", "epot", "etot", "press", "temp", "rdf"]
    
    for output in outputs:
        if step_type == "single_step":
            if output != "rdf":
                solid[output] = pd.read_csv("./"+step_type+"/solid/output_"+output+".dat", sep="\t")
                liquid[output] = pd.read_csv("./"+step_type+"/liquid/output_"+output+".dat", sep="\t")
                gas[output] = pd.read_csv("./"+step_type+"/gas/output_"+output+".dat", sep="\t")
        else:
            solid[output] = pd.read_csv("./"+step_type+"/solid/output_"+output+".dat", sep="\t")
            liquid[output] = pd.read_csv("./"+step_type+"/liquid/output_"+output+".dat", sep="\t")
            gas[output] = pd.read_csv("./"+step_type+"/gas/output_"+output+".dat", sep="\t")
                
    states = {
        "solid": solid,
        "liquid": liquid,
        "gas": gas
    }
        
    return states


def auto_correlation(x, i_max):
    A, B = [], []
    mean, mean2 = 0, 0
    
    for i in range(i_max):
        aux_A, aux_B1, aux_B2 = 0, 0, 0
        for j in range(i_max - i):
            aux_A += x[j] * x[j+i]
            aux_B1 += x[j]
            aux_B2 += x[j+i]
        A.append(aux_A/(i_max - i))
        B.append(aux_B1*aux_B2/(i_max - i)**2)
        mean += x[i]
        mean2 += x[i]**2
    
    A = np.array(A)
    B = np.array(B)
    mean = mean/i_max
    mean2 = mean2/i_max
    
    return (A - B)/(mean2 - mean**2)


def error(AV, AV2, n):
    if n == 0:
        return 0
    else:
        return np.sqrt((AV2[n] - AV[n]**2)/n)


def block_mean(x, L, last_only=True):
    M = len(x)
    if M%L != 0:
        raise ValueError("Lenght of the array x must be divisible by L")
    N = int(M/L)
    ave, ave2 = np.zeros(N), np.zeros(N)
    sum_prog, sum2_prog, err_prog = np.zeros(N), np.zeros(N), np.zeros(N)
    
    for i in range(N):
        summ = 0
        for j in range(L):
            k = j+i*L
            summ += x[k]
        ave[i] = summ/L
        ave2[i] = (ave[i])**2

    for i in range(N):
        for j in range(i+1):
            sum_prog[i] += ave[j]
            sum2_prog[i] += ave2[j]
        sum_prog[i] /= (i+1)
        sum2_prog[i] /= (i+1)
        err_prog[i] = error(sum_prog, sum2_prog, i)
    
    if last_only:
        return err_prog[-1]
    else:
        return err_prog
```

### Esercizio 07.2

```python
# loading data
states = load_data("single_step")
```

```python
# autocorrelation function
titles = ["solid", "liquid", "gas"]
colors = ["red", "blue", "green"]
i_max = int(1./5. * len(states["solid"]["epot"]))

plt.rcParams.update({'font.size':12})
fig, ax = plt.subplots(figsize=(8, 5))
fig.suptitle("Autocorrelation of U/N as a function of the number of steps")

for i in range(3):
    title = titles[i]
    color = colors[i]
    autocorr = auto_correlation(np.array(states[title]["epot"]["Mean"]), i_max)
    steps = np.arange(len(autocorr))
    
    ax.plot(steps, autocorr, color=color, label=title)
    ax.set_ylabel(r"$\chi$")
    ax.set_xlabel("step")
    ax.legend()
    ax.grid()
```

```python
# statistical uncertainty as a function of L
Ls = np.array([10, 20, 50, 100, 200, 500, 1000, 2000, 5000])
titles = ["solid", "liquid", "gas"]

plt.rcParams.update({'font.size':12})
fig, ax = plt.subplots(figsize=(8, 5))
fig.suptitle("Statistical uncertainty as a function of L (number of steps per block)")

for i in range(3):
    title = titles[i]
    state = states[title]
    errs = []
    for L in Ls:
        errs.append(block_mean(state["epot"]["Mean"], L))
    
    ax.plot(Ls, errs, label=title)
    ax.set_xscale("log")
    ax.set_ylabel("statistical uncertainty")
    ax.set_xlabel("L")
    ax.set_ylim([0., 0.005])
    ax.legend()
    ax.grid()
```

### Esercizio 07.4

```python
states_multi = load_data("multi_step")
states_MD = load_data("MD")
```

```python
# plot of U/N, P and g(r) for MC and MD simulations
titles = ["solid", "liquid", "gas"]

plt.rcParams.update({'font.size':13})
fig = plt.figure(figsize=(25, 17))
fig.suptitle("U/N, P and g(r) for MC and MD simulations")

for i in range(3):
    title = titles[i]
    state_multi = states_multi[title]
    state_MD = states_MD[title]
    
    ax = plt.subplot(3, 3, 1+3*i)
    ax.errorbar(state_multi["epot"]["Block"], state_multi["epot"]["Prog_mean"], yerr=state_multi["epot"]["Error"], color="g", label="MC")
    ax.errorbar(state_MD["epot"]["Block"], state_MD["epot"]["Prog_mean"], yerr=state_MD["epot"]["Error"], color="b", label="MD")
    ax.set_ylabel("U/N")
    ax.set_xlabel("blocks")
    ax.legend()
    ax.grid()
    
    ax = plt.subplot(3, 3, 2+3*i)
    ax.set_title(title)
    ax.errorbar(state_multi["press"]["Block"], state_multi["press"]["Prog_mean"], yerr=state_multi["press"]["Error"], color="g", label="MC")
    ax.errorbar(state_MD["press"]["Block"], state_MD["press"]["Prog_mean"], yerr=state_MD["press"]["Error"], color="b", label="MD")
    ax.set_ylabel("P")
    ax.set_xlabel("blocks")
    ax.legend()
    ax.grid()
    
    ax = plt.subplot(3, 3, 3+3*i)
    ax.errorbar(state_multi["rdf"]["r"], state_multi["rdf"]["<g>"], yerr=state_multi["rdf"]["err_g"], color="g", label="MC")
    ax.errorbar(state_MD["rdf"]["r"], state_MD["rdf"]["<g>"], yerr=state_MD["rdf"]["err_g"], color="b", label="MD")
    ax.set_ylabel("<g(r)>")
    ax.set_xlabel("r")
    ax.legend()
    ax.grid()
```

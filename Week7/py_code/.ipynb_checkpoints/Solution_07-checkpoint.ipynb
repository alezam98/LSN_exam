### Abstract
As in lecture 4, we want to study the properties of a system of $108$ particles of Argon in an isolated system (microcanonical ensemble). The tasks are the following:
1. performing Monte Carlo simulations, we want to study the autocorrelation function of the istantaneous potential energy per particle $U/N$. Then, we will study the error (on the same observable) when performing the block mean as a function of the number of steps per block;
2. again, using MC simulations, we want to study the system in the three phases of matter (solid, liquid and gas), measuring the potential energy per particle $U/N$, the pressure $P$ and radial distribution function $g(r)$ in the interval $r \in [0, L/2]$, where $L$ is the lenght of the side of the cube in which the system is confined;
3. lastly, we want to include the calculation of $g(r)$ in the MD code of Lecture 4. Then, using similar thermodynamics conditions, we want to compare the results obtained with it and the ones obtained with the MC simulations.

Before starting the simulation, we must include tail corrections for potential energy, $U$, and pressure $P$. Once we have done that, as always, the system must also be equilibrated, so that we can start correctly measuring the desired quantities.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_data(step_type):
    solid, liquid, gas = {}, {}, {}
    outputs = ["ekin", "epot", "etot", "press", "temp", "rdf"]
    
    for output in outputs:
        if step_type == "single_step":
            if output != "rdf":
                solid[output] = pd.read_csv("../data/"+step_type+"/solid/output_"+output+".dat", sep="\t")
                liquid[output] = pd.read_csv("../data/"+step_type+"/liquid/output_"+output+".dat", sep="\t")
                gas[output] = pd.read_csv("../data/"+step_type+"/gas/output_"+output+".dat", sep="\t")
        else:
            solid[output] = pd.read_csv("../data/"+step_type+"/solid/output_"+output+".dat", sep="\t")
            liquid[output] = pd.read_csv("../data/"+step_type+"/liquid/output_"+output+".dat", sep="\t")
            gas[output] = pd.read_csv("../data/"+step_type+"/gas/output_"+output+".dat", sep="\t")
                
    states = {
        "solid": solid,
        "liquid": liquid,
        "gas": gas
    }
        
    return states


def auto_correlation(x, i_max):
    A, B = [], []
    mean, mean2 = 0, 0
    
    for i in range(i_max):
        aux_A, aux_B1, aux_B2 = 0, 0, 0
        
        for j in range(i_max - i):
            aux_A += x[j] * x[j+i]
            aux_B1 += x[j]
            aux_B2 += x[j+i]
            
        A.append(aux_A / (i_max - i))
        B.append(aux_B1 * aux_B2 / (i_max - i)**2)
        mean += x[i]
        mean2 += x[i]**2
    
    A = np.array(A)
    B = np.array(B)
    mean = mean/i_max
    mean2 = mean2/i_max
    
    A = A[:int(i_max/2)]
    B = B[:int(i_max/2)]
    
    return (A - B)/(mean2 - mean**2)


def error(AV, AV2, n):
    if n == 0:
        return 0
    else:
        return np.sqrt((AV2[n] - AV[n]**2)/n)


def block_mean(x, L, last_only=True):
    M = len(x)
    N = int(M/L)
    ave, ave2 = np.zeros(N), np.zeros(N)
    sum_prog, sum2_prog, err_prog = np.zeros(N), np.zeros(N), np.zeros(N)
    
    for i in range(N):
        summ = 0
        for j in range(L):
            k = j+i*L
            summ += x[k]
        ave[i] = summ/L
        ave2[i] = (ave[i])**2

    for i in range(N):
        for j in range(i+1):
            sum_prog[i] += ave[j]
            sum2_prog[i] += ave2[j]
        sum_prog[i] /= (i+1)
        sum2_prog[i] /= (i+1)
        err_prog[i] = error(sum_prog, sum2_prog, i)
    
    if last_only:
        return err_prog[-1]
    else:
        return err_prog
```

### Exercise 07.2

```python
states = load_data("../data/single_step")
```

```python
titles = ["solid", "liquid", "gas"]
colors = ['orange', 'blue', 'green']
i_max = 4000


print("This will take a while...")
    
plt.rcParams.update({'font.size':13})
fig, ax = plt.subplots(figsize=(10, 6))
fig.suptitle("Autocorrelation of istantaneous U/N")

for i in range(3):
    title = titles[i]
    color = colors[i]
    autocorr = auto_correlation(np.array(states[title]["epot"]["Mean"]), i_max)
    steps = np.arange(len(autocorr))

    ax.plot(steps, autocorr, color=color, label=title, alpha=0.7)
    
ax.plot([0, int(i_max/2)], [0, 0], color="red")
ax.set_ylabel(r"$\chi$")
ax.set_xlabel("steps")
ax.legend()
ax.grid()
```

We can see that the autocorrelation function drops quite fast after approximately $250$ steps. Moreover, the oscillations are stronger changing the phase of the system, from solid to gas.

```python
Ls = np.logspace(1, 4, num=50, dtype=int)


print("This will also take a while...")

plt.rcParams.update({'font.size':13})
fig, ax = plt.subplots(figsize=(10, 6))
fig.suptitle("Statistical uncertainty as a function of L (number of steps per block)")

for i in range(3):
    title = titles[i]
    state = states[title]
    errs = []
    for L in Ls:
        errs.append(block_mean(state["epot"]["Mean"], L))

    ax.plot(Ls, errs, label=title)
    ax.set_xscale("log")
    ax.set_ylabel("err(U/N)")
    ax.set_xlabel("block length")
    ax.set_ylim([0., 0.005])
    ax.legend()
    ax.grid()
```

We can see that, when the system is in the solid or liquid stage, the error on the observable stabilizes once the size of the block reaches approximately $BL \approx 10^3$ steps. As in the autocorrelation function plot, the gaseous phase is once again the most unstable and the error starts stabilizing only for $BL \geq 5 \cdot 10^3$.

### Exercise 07.4

The next pictures report the values of $U/N$, $P$, and $g(r)$ computed performing MC simulations, in the solid, liquid and gaseous phases.

```python
states_multi = load_data("../data/multi_step")
states_MD = load_data("../data/MD")
```

#### Solid phase

```python
plt.rcParams.update({'font.size':13})
fig = plt.figure(figsize=(10, 16))
fig.suptitle("MC simulation, solid phase", y=0.93)

title = "solid"
state_multi = states_multi[title]

ax = plt.subplot(3, 1, 1)
ax.errorbar(state_multi["epot"]["Block"], state_multi["epot"]["Prog_mean"], yerr=state_multi["epot"]["Error"])
ax.set_ylabel(r"U/N [$\epsilon$]")
ax.set_xlabel("blocks")
ax.grid()

ax = plt.subplot(3, 1, 2)
ax.errorbar(state_multi["press"]["Block"], state_multi["press"]["Prog_mean"], yerr=state_multi["press"]["Error"])
ax.set_ylabel(r"P [$\epsilon\,m/\sigma^3$]")
ax.set_xlabel("blocks")
ax.grid()

ax = plt.subplot(3, 1, 3)
ax.errorbar(state_multi["rdf"]["r"], state_multi["rdf"]["<g>"], yerr=state_multi["rdf"]["err_g"])
ax.set_ylabel("<g(r)>")
ax.set_xlabel("r")
ax.grid()

plt.show()


print("On the last block")
print(f"(U/N)  estimate: {format(state_multi['epot']['Prog_mean'][99], '.5f')}  {format(state_multi['epot']['Error'][99], '.5f')}")
print(f"(P)    estimate: {format(state_multi['press']['Prog_mean'][99], '.5f')}  {format(state_multi['press']['Error'][99], '.5f')}")
```

#### Liquid phase

```python
plt.rcParams.update({'font.size':13})
fig = plt.figure(figsize=(10, 16))
fig.suptitle("MC simulation, liquid phase", y=0.93)

title = "liquid"
state_multi = states_multi[title]

ax = plt.subplot(3, 1, 1)
ax.errorbar(state_multi["epot"]["Block"], state_multi["epot"]["Prog_mean"], yerr=state_multi["epot"]["Error"])
ax.set_ylabel(r"U/N [$\epsilon$]")
ax.set_xlabel("blocks")
ax.grid()

ax = plt.subplot(3, 1, 2)
ax.errorbar(state_multi["press"]["Block"], state_multi["press"]["Prog_mean"], yerr=state_multi["press"]["Error"])
ax.set_ylabel(r"P [$\epsilon\,m/\sigma^3$]")
ax.set_xlabel("blocks")
ax.grid()

ax = plt.subplot(3, 1, 3)
ax.errorbar(state_multi["rdf"]["r"], state_multi["rdf"]["<g>"], yerr=state_multi["rdf"]["err_g"])
ax.set_ylabel("<g(r)>")
ax.set_xlabel("r")
ax.grid()

plt.show()


print("On the last block")
print(f"(U/N)  estimate: {format(state_multi['epot']['Prog_mean'][99], '.5f')}  {format(state_multi['epot']['Error'][99], '.5f')}")
print(f"(P)    estimate: {format(state_multi['press']['Prog_mean'][99], '.5f')}  {format(state_multi['press']['Error'][99], '.5f')}")
```

#### Gaseous phase

```python
plt.rcParams.update({'font.size':13})
fig = plt.figure(figsize=(10, 16))
fig.suptitle("MC simulation, gaseous phase", y=0.93)

title = "gas"
state_multi = states_multi[title]

ax = plt.subplot(3, 1, 1)
ax.errorbar(state_multi["epot"]["Block"], state_multi["epot"]["Prog_mean"], yerr=state_multi["epot"]["Error"])
ax.set_ylabel(r"U/N [$\epsilon$]")
ax.set_xlabel("blocks")
ax.grid()

ax = plt.subplot(3, 1, 2)
ax.errorbar(state_multi["press"]["Block"], state_multi["press"]["Prog_mean"], yerr=state_multi["press"]["Error"])
ax.set_ylabel(r"P [$\epsilon\,m/\sigma^3$]")
ax.set_xlabel("blocks")
ax.grid()

ax = plt.subplot(3, 1, 3)
ax.errorbar(state_multi["rdf"]["r"], state_multi["rdf"]["<g>"], yerr=state_multi["rdf"]["err_g"])
ax.set_ylabel("<g(r)>")
ax.set_xlabel("r")
ax.grid()

plt.show()


print("On the last block")
print(f"(U/N)  estimate: {format(state_multi['epot']['Prog_mean'][99], '.5f')}  {format(state_multi['epot']['Error'][99], '.5f')}")
print(f"(P)    estimate: {format(state_multi['press']['Prog_mean'][99], '.5f')}  {format(state_multi['press']['Error'][99], '.5f')}")
```

```python
plt.rcParams.update({'font.size':13})
fig = plt.figure(figsize=(14, 20))
fig.suptitle(r"MC-MD simulations, Radial Distance Function comparison", y=0.93)

for i in range(len(titles)):
    title = titles[i]
    state_multi = states_multi[title]
    state_MD = states_MD[title]

    ax = plt.subplot(3, 1, i+1)
    ax.set_title(title+" phase")
    ax.errorbar(state_multi["rdf"]["r"], state_multi["rdf"]["<g>"], yerr=state_multi["rdf"]["err_g"], color="g", label="MC")
    ax.errorbar(state_MD["rdf"]["r"], state_MD["rdf"]["<g>"], yerr=state_MD["rdf"]["err_g"], color="b", label="MD")
    ax.set_ylabel("<g(r)>")
    ax.set_xlabel("r")
    ax.legend()
    ax.grid()
```

First of all, there seems to be good agreement with the results gathered using MC and MD simulations. Moreover, we can observe the different shapes of the radial distance function in the three stages. In particular, we can see that the number of peaks decreases and the shape becomes more smooth as the system gets more chaotic and less ordered (i.e. switching from the solid to the liquid and then to the gas phase).

### Abstract
In this lecture, we want to parallelize with MPI libraries the Genetic Algorithm code previously implemented, in order to solve the TSP by performing a Genetic Search with different parallel GA searches of the optimal path. Each of the 4 nodes will perform an independent GA search, the so-called Continent, but every $N_{migr} = 10$ generations the Continents will exchange some their best individuals randomly.

Once we parallelize the code:
1. we will test it on the previous TSP, hence on 34 cities extracted on a circumference of radius $r = 1$ and inside a square of size length $l = 1$;
2. then we will execute the same code on the 50 american capitals, where the $(x, y)$ coordinates are defined by the latitude and the longitude respectively.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# global
processes = 4

def load_single_process_data(type_, process):
    dict_ = {}
    dict_ = {
        "cities": pd.read_csv("../data/"+type_+"/"+process+"_cities.dat".format(), sep="\t"),
        "chromosome": pd.read_csv("../data/"+type_+"/"+process+"_chromosome.dat", sep="\t"),
        "fitness": pd.read_csv("../data/"+type_+"/"+process+"_fitness.dat", sep="\t"),
        "mean_L": pd.read_csv("../data/"+type_+"/"+process+"_mean_L.dat", sep="\t")
    }
    return dict_


def load_data(city_type, shape=None):
    dicts = []
    for i in range(processes):
        process = f"r{i}"
        
        if city_type == "geometrical":
            dicts.append(load_single_process_data(shape, process))
        
        elif city_type == "capitals":
            dicts.append(load_single_process_data(city_type, process))
        
    return dicts       


def plot_all(dicts, city_type, shape=None):
        
    colors = ["blue", "green", "orange", "red"]
        
    plt.rcParams.update({'font.size':11})
    fig, axs = plt.subplots(figsize=(17, 8), ncols=4, nrows=2)
    gs = axs[1, 2].get_gridspec()
    for i in [0, 1]:
        for j in [2, 3]:
            axs[i, j].remove()
    axbig = fig.add_subplot(gs[0:, 2:])

    if city_type == "geometrical":
        if shape == "circle":
            fig.suptitle("Circle disposition", y=0.97)
        elif shape == "square":
            fig.suptitle("Square disposition", y=0.97)
    elif city_type == "capitals":
        fig.suptitle("American capitals", y=0.97)
    
    # trajectories
    for i in [0, 1]:
        for j in [0, 1]:
            ax = axs[i, j]
            p = 2*i + j #process number
            color = colors[p]

            ax.plot(dicts[0]["cities"]["city_x"], dicts[0]["cities"]["city_y"], "ro", label="cities")
            ax.plot(dicts[0]["cities"]["city_x"][0], dicts[0]["cities"]["city_y"][0], "bo", label="first city")
            ax.plot(dicts[p]["chromosome"]["city_x"], dicts[p]["chromosome"]["city_y"], color = color,
                    linestyle="--", lw=1.5, label=f"process {p}")

            if city_type == "geometrical":
                if shape == "circle":
                    theta = np.linspace(0, np.pi, 200)
                    ax.plot(np.cos(theta), np.sin(theta), color="grey")
                    ax.plot(np.cos(theta), -np.sin(theta), color="grey")
                elif shape == "square":
                    ax.plot([0, 0], [0, 1], color="grey")
                    ax.plot([0, 1], [1, 1], color="grey")
                    ax.plot([1, 1], [1, 0], color="grey")
                    ax.plot([1, 0], [0, 0], color="grey")

            if city_type == "geometrical":
                ax.set_xlabel("x")
                ax.set_ylabel("y")
            elif city_type == "capitals":
                ax.set_xlabel("latitude")
                ax.set_ylabel("longitude")
            ax.legend()
    
    # <L>
    ax = axbig
    gens = np.arange(0, len(dicts[0]["mean_L"]))
    for i in range(len(dicts)):
        dic = dicts[i]
        color = colors[i]
        #ax.errorbar(gens, dic["mean_L"]["Mean"], yerr=dic["mean_L"]["Error"], label=f"process {i}")
        ax.plot(gens, np.ones(len(dic["mean_L"]))*dic["chromosome"]["fitness"][0], color=color,
                linestyle="--", label=f"process {i}, final L")
        ax.plot(dic["fitness"]["gen"], dic["fitness"]["fitness"], color=color, label=f"process {i}, best L")

    ax.set_xlabel("generation")
    ax.set_ylabel("<L>")
    ax.legend()
    
    plt.show()
    
    
    for i in range(len(dicts)):
        print(f"Process {i}, best L: ", dicts[i]["chromosome"]["fitness"][0])
    print("")
    
    
def plot_best(dicts, city_type, shape=None):
    
    l = []
    for i in range(len(dicts)):
        l.append(dicts[i]["chromosome"]["fitness"][0])
    best_p, = np.where( np.array(l) == np.min( np.array(l) ) )
    best_p = best_p[0]
    dic = dicts[best_p]
        
    plt.rcParams.update({'font.size':11})
    fig, (ax1, ax2) = plt.subplots(figsize=(17, 8), ncols=2, nrows=1)
    
    if city_type == "geometrical":
        if shape == "circle":
            fig.suptitle(f"Circle disposition, best result (process {best_p})", y=0.97)
        elif shape == "square":
            fig.suptitle(f"Square disposition, best result (process {best_p})", y=0.97)
    elif city_type == "capitals":
        fig.suptitle(f"American capitals, best result (process {best_p})", y=0.97)
    
    # trajectories
    ax = ax1
    ax.plot(dic["cities"]["city_x"], dic["cities"]["city_y"], "ro", label="cities")
    ax.plot(dic["cities"]["city_x"][0], dic["cities"]["city_y"][0], "bo", label="first city")
    ax.plot(dic["chromosome"]["city_x"], dic["chromosome"]["city_y"], color = "green",
            linestyle="--", lw=1.5, label=f"best trajectory")

    if city_type == "geometrical":
        if shape == "circle":
            theta = np.linspace(0, np.pi, 200)
            ax.plot(np.cos(theta), np.sin(theta), color="grey")
            ax.plot(np.cos(theta), -np.sin(theta), color="grey")
        elif shape == "square":
            ax.plot([0, 0], [0, 1], color="grey")
            ax.plot([0, 1], [1, 1], color="grey")
            ax.plot([1, 1], [1, 0], color="grey")
            ax.plot([1, 0], [0, 0], color="grey")

    if city_type == "geometrical":
        ax.set_xlabel("x")
        ax.set_ylabel("y")
    elif city_type == "capitals":
        ax.set_xlabel("latitude")
        ax.set_ylabel("longitude")
    ax.legend()
    
    # <L>
    ax = ax2
    gens = np.arange(0, len(dicts[0]["mean_L"]))
    ax.errorbar(gens, dic["mean_L"]["Mean"], yerr=dic["mean_L"]["Error"], color="blue", label=f"<L>")
    ax.plot(gens, np.ones(len(dic["mean_L"]))*dic["chromosome"]["fitness"][0], color="red",
            linestyle="--", label=f"final L")
    ax.plot(dic["fitness"]["gen"], dic["fitness"]["fitness"], color="green", label=f"best L")

    ax.set_xlabel("generation")
    ax.set_ylabel("<L>")
    ax.legend()
    
    #fig.tight_layout()
```

### Exercise 10.1

```python
city_type = "geometrical"
shape = "circle"
dicts_circle = load_data(city_type, shape)
```

```python
plot_all(dicts_circle, city_type, shape);
```

First of all, we test our parallelized Genetic Algorithm code on the easiest problem, the 34 cities on a circumference. We can see that, as in the single process case analyzed in the previous lecture, every Continent (i.e. each individual node) converges to the optimal solution.

An interesting observation could be made looking at the graph on the right, where the best sequence loss function is reported for each generation and node. We can see that at the 30th generation, there is a sudden increase in the best sequence loss value. This is due to the degeneracy of the problem: fixing the starting city, the best path can be run across clockwise and anticlockwise, hence there are actually two distincted best paths to which the nodes can converge. In this case, at the 30th generation, when the 3rd migration occured, nodes converging to different best paths exchanged their best sequences, which caused the sudden increase in the loss function value.

Since, once again, the algorithm worked fine on the easy problem, we can now test it on more difficult ones.

```python
city_type = "geometrical"
shape = "square"
dicts_square = load_data(city_type, shape)
```

```python
plot_all(dicts_square, city_type, shape)
plot_best(dicts_square, city_type, shape);
```

Once we test the code on a set of cities inside a square, we note that, in this case, the different processes converge to different sequences, i.e. local minima of the loss/cost function. We can see that, out of the 4 Continents, the best path has been found by process 1. This shows what anticipated during the previous lecture analysis, that is the fact that we don't know if the algorithm converged to the actual best path.

In the top images, as in the previous example, we can see the best trajectories (on the left) and the best sequence loss value as a function of the generation (on the right). At the bottom, instead, the best process results are shown, i.e. the best trajectory (on the left) and the value of $\left\langle L \right\rangle$ calculated on the whole population, together with the best sequence fitness (on the right).

### Exercise 10.2

```python
city_type = "capitals"
dicts_capit = load_data(city_type)
```

```python
plot_all(dicts_capit, city_type)
plot_best(dicts_capit, city_type);
```

As in the previous example, when tested on the 50 amercan capitals positions, the different processes converge to different sequences. This result was definetely expected, since already with just 34 cities, the nodes were incapable to converge to the same result.

We can see that, out of the 4 Continents, the best path has been found by process 0. Again in the top images, we can see the best trajectories (on the left) and the best sequence loss value as a function of the generation (on the right). In the bottom images, instead, the best process results are shown, i.e. the best trajectory (on the left) and the value of $<L>$ calculated on the whole population, together with the best sequence fitness (on the right).

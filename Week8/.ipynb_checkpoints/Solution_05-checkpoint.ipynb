### Pacchetti

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

### Esercizio 05.1

```python
# caricamento dati
config_dfs = {
    "gs": pd.read_csv("./data_05/configs_gs.txt", sep="\t"),
    "1e": pd.read_csv("./data_05/configs_1e.txt", sep="\t")
}

mean_dfs = {
    "gs_uniform": pd.read_csv("./data_05/r_gs_uniform.txt", sep="\t"),
    "1e_uniform": pd.read_csv("./data_05/r_1e_uniform.txt", sep="\t"),
    "gs_gauss": pd.read_csv("./data_05/r_gs_gauss.txt", sep="\t"),
    "1e_gauss": pd.read_csv("./data_05/r_1e_gauss.txt", sep="\t"),
}
```

```python
# plot delle medie
blocks = np.arange(len(mean_dfs["gs_uniform"])) + 1
mean_gs = 1.5
mean_1e = 5.

plt.rcParams.update({'font.size':12})
fig, (ax1, ax2) = plt.subplots(figsize=(16, 5), nrows=1, ncols=2)

# ... per il ground state...
ax1.set_title("Ground state: |n, m, l> = |1, 0, 0>")
ax1.errorbar(blocks, mean_dfs["gs_uniform"]["Mean"], yerr=mean_dfs["gs_uniform"]["Error"], color="b", label="Uniform")
ax1.errorbar(blocks, mean_dfs["gs_gauss"]["Mean"], yerr=mean_dfs["gs_gauss"]["Error"], color="g", label="Gauss")
ax1.plot([blocks.min(), blocks.max()], [mean_gs, mean_gs], color="r", label="Teorical")
ax1.set_xlabel("blocks")
ax1.set_ylabel(r"<r>/a$_0$")
ax1.legend()
ax1.grid()

# ... e per il primo stato eccitato
ax2.set_title("First excited state: |n, m, l> = |2, 1, 0>")
ax2.errorbar(blocks, mean_dfs["1e_uniform"]["Mean"], yerr=mean_dfs["1e_uniform"]["Error"], color="b", label="Uniform")
ax2.errorbar(blocks, mean_dfs["1e_gauss"]["Mean"], yerr=mean_dfs["1e_gauss"]["Error"], color="g", label="Gauss")
ax2.plot([blocks.min(), blocks.max()], [mean_1e, mean_1e], color="r", label="Teorical")
ax2.set_xlabel("blocks")
ax2.set_ylabel(r"<r>/a$_0$")
ax2.legend()
ax2.grid()
```

```python
# plot delle configurazioni
df_gs = config_dfs["gs"]
df_1e = config_dfs["1e"]

plt.rcParams.update({'font.size':12})
fig = plt.figure(figsize=(16, 5))

# ... per il ground state...
ax1 = plt.subplot(1, 2, 1, projection="3d")
ax1.set_title("Ground state: |n, m, l> = |1, 0, 0>")
ax1.scatter(df_gs["x"], df_gs["y"], df_gs["z"], c=df_gs["z"], marker=".")
ax1.set_xlabel(r'x/a$_0$')
ax1.set_ylabel(r'y/a$_0$')
ax1.set_zlabel(r'z/a$_0$')
ax1.view_init(10, 30)

# ... e per il primo stato eccitato
ax2 = plt.subplot(1, 2, 2, projection="3d")
ax2.set_title("First excited state: |n, m, l> = |2, 1, 0>")
ax2.scatter(df_1e["x"], df_1e["y"], df_1e["z"], c=df_1e["z"], marker=".")
ax2.set_xlabel(r'x/a$_0$')
ax2.set_ylabel(r'y/a$_0$')
ax2.set_zlabel(r'z/a$_0$')
ax2.view_init(10, 30)
```
